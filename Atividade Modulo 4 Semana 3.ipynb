```python
import pytest

def calcular_valor_com_desconto(valor_unitario, quantidade):
    desconto = 1

    if quantidade >= 10 and quantidade <= 99:
        desconto = 0.95
    elif quantidade >= 100 and quantidade <= 999:
        desconto = 0.90
    elif quantidade >= 1000:
        desconto = 0.85

    valor_com_desconto = valor_unitario * desconto * quantidade
    valor_sem_desconto = valor_unitario * quantidade

    return valor_sem_desconto, valor_com_desconto

# função de testes
def test_desconto():
    assert calcular_valor_com_desconto(10, 5) == (50, 50) # testando se q quantidade é menor que 10
    assert calcular_valor_com_desconto(10, 10) == (100, 95) # testando se a quantidade está entre 10 e 99
    assert calcular_valor_com_desconto(10, 100) == (1000, 900) # testando se a quqntidade está entre 100 e 999
    assert calcular_valor_com_desconto(10, 1000) == (10000, 8500) # testando se a quantidade é >= a 1000
```

```python
import pytest
from unittest.mock import patch

pecas = []

# Função que gera o código para os itens
def gerar_codigo():
    tamanho_lista = len(pecas)
    if tamanho_lista > 0:
        ultima_peca = pecas[tamanho_lista - 1]
        ultimo_codigo = ultima_peca['codigo']
        return ultimo_codigo + 1

    return 1

# Função de cadastro de itens
def cadastrar_peca(codigo, nome, fabricante, valor):
    peca = {
        'codigo': codigo,
        'nome': nome,
        'fabricante': fabricante,
        'valor': valor
    }

    pecas.append(peca)
    return peca

# Função que remove itens
def remover_peca(codigo):
    peca_remover = {}
    for peca in pecas:
        if peca['codigo'] == codigo:
            peca_remover = peca
            break

    if peca_remover:
        pecas.remove(peca_remover)
        return True

    return False

# Testes unitários para as funções

def test_gerar_codigo(): # Passou no Teste
    assert gerar_codigo() == 1
    cadastrar_peca(1, 'Peca 1', 'Fabricante 1', 10.0)
    assert gerar_codigo() == 2

def test_cadastrar_peca(): # Passou no teste
    peca = cadastrar_peca(2, 'Peca 2', 'Fabricante 2', 20.0)
    assert peca in pecas

def test_remover_peca(): 
    assert remover_peca(2) == True
    assert remover_peca(2) == False
    #Erro a função da remover_peca está retornando o valor boleano "False" ao invés de "True".
```
